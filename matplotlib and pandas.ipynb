# Groupby 

```python
import pandas as pd
```

```python
df = pd.read_csv("weather_by_cities.csv")
df
```

```python
g = df.groupby("city")
g
```

```python
for city, city_df in g:#groupby function gives index and other data so here we declare two thing is city and city_df
    print(city)
    print(city_df)
```

```python
g.get_group("mumbai")
```

```python
g.max()
```

```python
g.mean()
```

```python
g.describe()
```

```python
%matplotlib inline
g.plot()
```

# crosstab 

```python
import pandas as pd
```

```python
df =pd.read_excel("survey.xls",index_col = 0)
df
```

```python
pd.crosstab([df.Sex],[df.Handedness])#margins  = true give total of frequency distribution
```

```python
pd.crosstab([df.Sex],[df.Handedness],normalize='index')
```

```python
pd.crosstab([df.Sex],[df.Handedness],values=df.Age,aggfunc=np.average)
```
